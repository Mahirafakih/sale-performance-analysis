```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../data/sales_data_2023.csv')
df.head()
```

```python
print(df.isnull().sum())
```

```python
# Fill missing categorical values with mode
df['Region'].fillna(df['Region'].mode()[0], inplace=True)
df['Product'].fillna(df['Product'].mode()[0], inplace=True)
df['Salesperson'].fillna(df['Salesperson'].mode()[0], inplace=True)
```

```python
print(df.info())

print(df.describe())

# Total revenue
total_revenue = df['Total_Sales'].sum()
print(f"Total Revenue: ₹{total_revenue:,.0f}")
```

```python
#Top 5 products by revenue
top_products = df.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,6))
sns.barplot(x=top_products.values, y=top_products.index)
plt.title('Top 5 Products by Revenue')
plt.xlabel('Total Sales (₹)')
plt.tight_layout()
plt.savefig('../visuals/top_products.png')
plt.show()
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Month'] = df['Order_Date'].dt.to_period('M')

monthly_sales = df.groupby('Month')['Total_Sales'].sum()

plt.figure(figsize=(10,5))
monthly_sales.plot(marker='o')
plt.title('Monthly Sales Trend (2023)')
plt.ylabel('Total Sales (₹)')
plt.xlabel('Month')
plt.grid(True)
plt.tight_layout()
plt.savefig('../visuals/monthly_sales_trend.png')
plt.show()
```

```python
region_sales = df.groupby('Region')['Total_Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(x=region_sales.index, y=region_sales.values)
plt.title('Total Sales by Region')
plt.ylabel('Total Sales (₹)')
plt.xlabel('Region')
plt.tight_layout()
plt.savefig('../visuals/revenue_by_region.png')
plt.show()
```

```python
print("Key Insights:")
print(f"- Total Revenue Generated: ₹{df['Total_Sales'].sum():,.0f}")
print(f"- Top Region: {region_sales.idxmax()}")
print(f"- Best Selling Product: {top_products.idxmax()}")
```
